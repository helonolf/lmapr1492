Diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

Loi de Bragg : $n \lambda = 2d \sin(\theta)$ 

--> d est la distance entre deux plans (h k l)

--> $2\theta$ est l'angle entre les directions d’incidence et d’observation

--> n est l'ordre de la réflexion

--> $\lambda$ est la longueur d'onde du rayonnement X incident (ici : λ=1.54060 Å)

Pour les trois premier pics, les indices (hkl), $2\theta$, d et l'intensité sont calculé ci-dessous.
